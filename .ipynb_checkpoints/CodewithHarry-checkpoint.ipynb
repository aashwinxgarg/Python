```python
#tut8    String Slicing

mystr="Harry is a good boy"
print(mystr)
#H is 0, a is 1, r is 2, r is 3, y is 4,  is 5, i is 6, s is 7, etc,.
print(mystr[4])
print(mystr[5])
print(mystr[0:4])    #complete Harry was not printed because 0 is included and 4 is excluded in this, if 4 would have been included, then Harry would have been printed.
print(mystr[0:5])    #here, i wrote 5 to include letter 4 i.e. y and 5 i.e. space is excluded which makes the word Harry complete.
print(len(mystr))    #index is from 0 to 18 which makes it a total of 19.
print(mystr[0:18])    #y is excluded here because of 18.
print(mystr[0:19])    #y included.
print(mystr[0:78])
print(mystr[0:5:2])
print(mystr[0:])
print(mystr[:5])
print(mystr[:])
print(mystr[::])    #if it is [x:y:z], then, by default, x is 0, y is length of string and z is 1.
print(mystr[0:19:1])
print(mystr[:19:1])
print(mystr[::1])
print(mystr[::2])
print(mystr[::3])
print(mystr[::9])
print(mystr[::5559])
print(mystr[-1:0])
print(mystr[-4:0])
print(mystr[-4:])
print(mystr[-4:-2])
print(mystr[15:17])    #same as the upper result because replaced negatives in poitives but it still is the same position as in mystr: 15 in positive replaces -4 in negative, and in the same way, 17 in the positive replaces -2 in negative.
print(mystr[::-1])    #it will invert the whole string.
print(mystr[::-2])
print(type(mystr))
print(mystr.isalnum())
mystr= "Harryisagoodboy"
print(mystr.isalnum())
mystr= "Harryisagoodboy69"
print(mystr.isalnum())
mystr="Harry is a good boy"
print(mystr.isalpha())
mystr= "Harryisagoodboy"
print(mystr.isalpha())
mystr= "Harryisagoodboy69"
print(mystr.isalpha())
mystr= "Harry is a good boy"
print(mystr.endswith("boy"))
print(mystr.endswith("oy"))
print(mystr.endswith("y"))
print(mystr.endswith("bdoy"))
print(mystr.count("a"))
print(mystr.count("H"))
print(mystr.count("o"))
print(mystr.count("b"))
mystr="harry is a good boy"
print(mystr.capitalize())
print(mystr.find("is"))
print(mystr.find("o"))
print(mystr.upper())
mystr="HARRY IS A GOOD BOY"
print(mystr.lower())
mystr="Harry is a good boy"
print(mystr.replace("is", "are"))
print(mystr.replace("good", "bad"))
```

```python
#tut9    Lists and lists functions
grocery= ["Harpic", "vim bar", "deodrant", "Bhindi", "Lollypop", 56]
print(grocery)
print(grocery[0])
print(grocery[1])
print(grocery[5])
#print(grocery[6])    showed error because there was no number 6 in the list formed until now
numbers= [2, 7, 9, 11, 3]
print(numbers)
print(numbers[2])
numbers.sort()
print(numbers)
numbers.reverse()
print(numbers)
#print(numbers[9])    error because there is no number 6 in the list formed till now
numbers=[2, 7, 9, 11, 3]
print(numbers[0:5])
print(numbers[:5])
print(numbers[:])
print(numbers[1:])
print(numbers[1:4])
print(numbers)    #The old list does not change due to slicing function, but sort and reverse function do make a change in the list.
print(numbers[::1])
print(numbers[::2])
print(numbers[::3])
print(numbers[::-1])     #It is preferred not to take any negative sign in slicing and not at al less than -1.
print(numbers[::-2])     #It is preferred not to take any negative sign in slicing and not at al less than -1.
print(numbers[::-3])    #It is preferred not to take any negative sign in slicing and not at al less than -1.
#for example, you may observe the next instruction
print(numbers[1:5:-3])
print(numbers[1:5:2])
print(len(numbers))
print(max(numbers))
print(min(numbers))
numbers.append(7)    #we can add any amount of number and we can add any number, also we can add a single number repeatedly.
numbers.append(7)    #append means add it to the end of the list
numbers.append(55)
numbers.append(52)
print(numbers)
num=[]
num.append(71)
num.append(71)
num.append(71)
print(num)
nums=[]
nums.append(72)    #will add a number in the end of the list
nums.append(5)
nums.append(1)
print(nums)
numbers= [2, 7, 9, 11, 3]
numbers.insert(1,67)    #will insert a number on a position you give based on indexing. (index,number)
print(numbers)
numbers.remove(9)    #can remove any number you want
print(numbers)
numbers.pop()    #wil pop off the last number present in the list.
print(numbers)
numbers[1]=98    #changes your number which you want to change. numbers[position]=number needed after change.
#Just changed a number in the list which means the list can be changed and thus, lists are mutable.
print(numbers)
#Lists are mutable but tupples are not
# Mutable- Can Change
# Immutable- Cannot Change
tp=(1, 2, 3)
# tp[1]=8    cannot change because this is a tupple and tupple are not mutable i.e. they are immutable.
print(tp)
tp=(1)
print(tp)
#as you might be able to see the result, when we put a single element, we do not get it as a tupple but we can get it if we apply an extra comma(,). Observe in the next instruction.
tp=(1,)
print(tp)
a=1
b=8
a,b=b,a    #this is a python special feature to inter-change the values of two variables.
print(a,b)
temp=a
a=b    #this is a traditional method to swap the two variable values stored within them.
b=temp
print(a, b)
```

```python
#tut10    dictionary and functions
# Dictionary is nothing but key value pairs
d1={}
print(type(d1))
d2={"harry":"burger", "rohan":"fish", "skillf":"roti"}    #this is a dictionary which has harry, rohan and skillf as key and burger fish and roti as values of that keys.
print(d2["harry"])
print(d2["rohan"])
d2={"harry":"burger", "rohan":"fish", "skillf":"roti", "shubham":{"breakfast":"pasta", "lunch": "rice and rajma", "dinner": "dal roti"}}    
#this is an example of dictionary within another dictionary as we have a dictionary named "shubham" within a dictionary named "d2" present in the starting of the above of this very line.
print(d2["shubham"])
print(d2["shubham"]["lunch"])
#the value of the key in the dictionary can be a dictionary itself, tupple, list, or else, but the keys present are immutable i.e. the keys of the dictionary cannot be changed.
d2["ankit"]="junk food"    #we can add a key-value pair in the dictionary just like this.
print(d2)
d2["aurangzeb"]="kebabs"    #we just added one more key-file pair
print(d2)
d2[420]="kebabs"
print(d2)
del d2[420]    #we can delete a key-value pair like this.
print(d2)
d3=d2
#del d3["shubham"]
print(d2)
d3=d2.copy()
del d3["shubham"]
print(d3)
print(d2)
print(d3.get("shubham"))
print(d3.get("harry"))
d2.update({"leena":"maggie"})
print(d2)
print(d2.keys())    #prints key values of dictionaries
print(d2.items())    #prints key value pairs of dictionaries
```
